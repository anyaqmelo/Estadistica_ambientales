```python
def import_csv(file):
    import pandas as pd 
    return pd.read_csv(file)
```

```python
path = "Terminos_lagoon_TA_.csv"
```

```python
CO2Data = import_csv(path)
print(CO2Data.info())
```

```python
import seaborn as sns
import matplotlib.pylab as plt 

my_colors = ["#EB9AC0", "#B3EBF2"]

sns.set(font_scale=2, style="ticks", palette = my_colors)
```

Diagrama de Caja

```python
ax = sns.catplot(data=CO2Data,
                 x="area", y="ta_micromol_kg",
                 hue="season", col="estuary",
                 kind="box",
                 height=7, aspect=1)

ax.set(xlabel="", ylabel='TA ($\mu mol \; kg^{-1}$)')

ax.set_titles(col_template="{col_name}", row_template="{row_name}")

plt.setp(ax.legend.get_title(), fontsize=20)

sns.despine(top=False, right=False)

plt.show()
```

Violinplot por área, season and stuary

```python
ax = sns.catplot(
    data=CO2Data,
    x="area", 
    y="ta_micromol_kg",
    hue="season", 
    col="estuary",
    kind="violin",
    height=7, 
    aspect=1,
    split=True  
)

ax.set(xlabel="", ylabel='TA ($\mu mol \; kg^{-1}$)')

ax.set_titles(col_template="{col_name}", row_template="{row_name}")

plt.setp(ax.legend.get_title(), fontsize=20)

sns.despine(top=False, right=False)

plt.show()
```

Scatterplt: TA vs DIC por temporada

```python
ax = sns.relplot(
    data=CO2Data,
    x="dic_micromol_kg",
    y="ta_micromol_kg",
    hue="season",
    col="estuary",
    kind="scatter",
    height=7,
    aspect=1
)
ax.set(xlabel='DIC ($\mu mol \; kg^{-1}$)', ylabel='TA ($\mu mol \; kg^{-1}$)')

ax.set_titles(col_template="{col_name}")

plt.setp(ax.legend.get_title(), fontsize=20)

sns.despine(top=False, right=False)

plt.show()
```

Lineplot de TA vs Dic por temporada


```python
ax = sns.relplot(
    data=CO2Data,
    x="depth_m",
    y="temp_c",
    hue="area",
    col="estuary",
    kind="line",
    height=7,
    aspect=1,
    marker="o"
)
ax.set(xlabel='Profundidad (m)', ylabel='Temperatura (°C)')
ax.set_titles(col_template="{col_name}")
plt.setp(ax.legend.get_title(), fontsize=20)
sns.despine(top=False, right=False)
plt.show()
```
